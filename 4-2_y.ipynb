```python
from sklearn.datasets import load_diabetes
diabets = load_diabetes()
```

```python
type(diabets.data)
```

```python
import numpy as np
X = np.array(diabets.data)
Y = np.array(diabets.target)
```

```python
X.shape
```

```python
Y.shape
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0], Y)
plt.show()
```

```python
X = X[:,2]
```

```python
X.shape
```

```python
plt.scatter(X,Y)
```

```python
w=1.0
b= 1.0

for i in range(1,5):
    for x_i, y_i in zip(X,Y):
        y_hat = x_i * w + b
        err = y_i - y_hat
        w_rate = x_i
        w += w_rate *  err
        b += 1* err

print(w,b)
```

```python
plt.scatter(X,Y)
x1 = [-0.1, 0.15]
y1 = [x1[0] * w + b, x1[1] * w + b]
plt.plot(x1,y1)
plt.show()
```
